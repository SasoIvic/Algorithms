### Naložimo testne slike

```python
slika_rgb = pyplot.imread('./1024px-Colouring_pencils.jpg')
slika_rggb = pyplot.imread('./1024px-Colouring_pencils_RGGB.png')
slika_grbg = pyplot.imread('./1024px-Colouring_pencils_GRBG.png')
slika_gbrg = pyplot.imread('./1024px-Colouring_pencils_GBRG.png')
slika_bggr = pyplot.imread('./1024px-Colouring_pencils_BGGR.png')

# png se ob nalaganju pretvori v np.float32 tip, pretvori nazaj 
# v np.uint8
slika_rggb = np.uint8(slika_rggb*255)
slika_grbg = np.uint8(slika_grbg*255)
slika_gbrg = np.uint8(slika_gbrg*255)
slika_bggr = np.uint8(slika_bggr*255)
```

### Izvorna slika
Ta slika je bila uporabljena za generiranje testnih slik z bayer vzorcem. Predstavlja pravilno rekonstruirano sliko, katere velikost je enaka velikosti bayer vzorca. Je torej 2x večja od decimirane in enako velika kot interpolirana slika.

```python
pyplot.imshow(slika_rgb)
pyplot.title('izvorna slika (pravilna velikost in barve)');
```

### Testi z decimacijo
Vse slike so polovice velikosti izvorne slike.

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_rggb, 'RGGB', False)

pyplot.imshow(slika_rgb_res)
pyplot.title('RGGB demoziciranje, brez interpolacije');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_grbg, 'GRBG', False)

pyplot.imshow(slika_rgb_res)
pyplot.title('GRBG demoziciranje, brez interpolacije');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_gbrg, 'GBRG', False)

pyplot.imshow(slika_rgb_res)
pyplot.title('GBRG demoziciranje, brez interpolacije');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_bggr, 'BGGR', False)

pyplot.imshow(slika_rgb_res)
pyplot.title('BGGR demoziciranje, brez interpolacije');
```

### Testi z interpolacijo 
Vse te slike bi morale biti enake velikosti kot izvorna slika.

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_rggb, 'RGGB', True)

pyplot.imshow(slika_rgb_res)
pyplot.title('RGGB demoziciranje, interpolirano');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_grbg, 'GRBG', True)

pyplot.imshow(slika_rgb_res)
pyplot.title('GRBG demoziciranje, interpolirano');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_gbrg, 'GBRG', True)

pyplot.imshow(slika_rgb_res)
pyplot.title('GBRG demoziciranje, interpolirano');
```

```python
slika_rgb_res = bayer.bayer_v_rgb(slika_bggr, 'BGGR', True)

pyplot.imshow(slika_rgb_res)
pyplot.title('BGGR demoziciranje, interpolirano');
```
