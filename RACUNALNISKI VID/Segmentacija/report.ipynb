```python
def izrisi_rezultate(slika, konture, analiza):
    
    pyplot.figure()
    pyplot.imshow(slika, cmap='gray')
    pyplot.axis('off')    
    for n in range(len(konture)):
        pyplot.plot(konture[n][:, 1], konture[n][:, 0], 'b', linewidth=2.)
        cy, cx, l1, l2, theta = analiza[n, :]

        theta_rad = theta*np.pi/180.
        
        v1 = np.array([-sin(theta_rad), cos(theta_rad)])
        v2 = np.array([v1[1], -1*v1[0]])

        v1 *= l1
        v2 *= l2

        pyplot.arrow(cx, cy, v1[1], v1[0], width=5, facecolor='g')
        pyplot.arrow(cx, cy, v2[1], v2[0], width=5, facecolor='r')
```

```python
slika = imread('./primer1.png')[:, :, :3]
slika = np.uint8(slika.mean(2)*255.)

konture = segmentacija.poisci_konture(slika, 1000)
rezultati = segmentacija.analiziraj_konture(konture)

izrisi_rezultate(slika, konture, rezultati)
```
